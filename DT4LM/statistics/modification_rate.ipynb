```python
def cal_word_diff(w1, w2):
    # indices = set()
    # for i in range(min(len(w1), len(w2))):
    #     if w1[i] != w2[i]:
    #         indices.add(i)
    # return indices
    return max(len(set(w1) - (set(w1) & set(w2))), len(set(w2) - (set(w1) & set(w2))))
```

```python
def tokenize_columns(example):
    tokens = []
    for key, value in example.items():
        if key != "label":
            tokens.extend(value.split())
    return tokens
```

```python
from datasets import load_dataset
def cal_dataset_stat(path_to_differential, path_to_original):
    dataset_differential = load_dataset(path_to_differential, split='train')
    dataset_original = load_dataset(path_to_original, split='train')
    if (len(dataset_differential) != len(dataset_original)) or len(dataset_differential) == 0 or len(dataset_original) == 0:
        raise Exception("Mismatch!")
    print("Length of this dataset: ", len(dataset_differential))
    lst_percentages = []
    for i in range(len(dataset_differential)):
        original_text = tokenize_columns(dataset_original[i])
        result_text = tokenize_columns(dataset_differential[i])
        num_words_changed = cal_word_diff(original_text, result_text)
        try:
            change_rate = ((num_words_changed * 100.0) / len(original_text))
            lst_percentages.append(change_rate)
        except:
            print("An error has occurred for input index: ", i)
    lst_percentages = [p for p in lst_percentages if p > 0]
    stat = (sum(lst_percentages) / len(lst_percentages)) if lst_percentages else 0
    stat = round(stat, 2)
    return stat
```
