### BERTScore - BERT-Precision, BERT-Recall, BERT-F1

```python
from bert_score import score
```

```python
def evaluate_bertscore(candidates, references):
    """
    Calculate BERTScore for a list of candidate texts against reference texts.
    Args:
        candidates (list of str): Generated text.
        references (list of str): Human-written reference text.
    Returns:
        tuple: Mean Precision, Recall, and F1 scores across all texts.
    """
    P, R, F1 = score(candidates, references, lang="en", rescale_with_baseline=True)
    return P.mean().item(), R.mean().item(), F1.mean().item()
```

```python
# Example usage
candidates = ["This is a sample generated text.", "Another example generated text."]
references = ["This is a sample reference text.", "Another example reference text."]
precision, recall, f1 = evaluate_bertscore(candidates, references)
print(f"BERTScore Precision: {precision}, Recall: {recall}, F1: {f1}")
```

### BLEU

```python
from nltk.translate.bleu_score import sentence_bleu
import numpy as np
```

```python
def evaluate_bleu(candidates, references):
    scores = [sentence_bleu([ref], cand) for cand, ref in zip(candidates, references)]
    return np.mean(scores)
```

```python
candidates = ["This is a sample generated text.", "Another example generated text."]
references = ["This is a sample reference text.", "Another example reference text."]
bleu_scores = evaluate_bleu(candidates, references)
print("BLEU Score:", bleu_scores)
```

### METEOR

```python
from nltk.translate.meteor_score import meteor_score
import nltk
import numpy as np
```

```python
# nltk.download('wordnet')
# nltk.download('omw-1.4')
```

```python
def evaluate_meteor(candidates, references):
    """ Calculate METEOR scores for a list of references and hypotheses.
    Args:
        references (list of str): A list of reference texts (each text should have only one reference).
        hypotheses (list of str): A list of hypothesis texts.
    Returns:
        float: The average METEOR score across all text pairs.
    """
    scores = []
    for reference, candidate in zip(references, candidates):
        # Tokenize the reference and hypothesis
        reference_tokens = nltk.word_tokenize(reference)
        candidate_tokens = nltk.word_tokenize(candidate)
        
        # Calculate METEOR score
        score = meteor_score([reference_tokens], candidate_tokens)
        scores.append(score)
    
    return np.mean(scores)
```

```python
candidates = ["This is a sample generated text.", "Another example generated text."]
references = ["This is a sample reference text.", "Another example reference text."]
meteor_scores = evaluate_meteor(candidates, references)
print("METEOR Score:", meteor_scores)
```

### ROUGE

```python
from rouge import Rouge
import numpy as np
```

```python
def evaluate_rouge(candidates, references):
    """ Calculate the ROUGE scores for a list of generated texts against references.
    Args:
        hypotheses (list of str): Generated texts.
        references (list of str): Original reference texts.
    Returns:
        float: Average ROUGE score.
    """
    rouge = Rouge()
    scores = []
    for cand, ref in zip(candidates, references):
        # Ensure each reference is a list of strings for compatibility with the ROUGE package
        score = rouge.get_scores(cand, ref)
        # Extract the F1 score from the ROUGE-L metric
        rouge_l_f1 = score[0]['rouge-l']['f']
        scores.append(rouge_l_f1)
    
    return np.mean(scores)
```

```python
candidates = ["This is a sample generated text.", "Another example generated text."]
references = ["This is a sample reference text.", "Another example reference text."]
rouge_scores = evaluate_rouge(candidates, references)
print("ROUGE Score:", rouge_scores)
```

##### RTE Dataset

```python
def extract_texts_rte_premise(dataset_path):
    dataset = load_dataset(dataset_path)
    texts = dataset['train']['premise']
    return texts
```

```python
def extract_texts_rte_hypothesis(dataset_path):
    dataset = load_dataset(dataset_path)
    texts = dataset['train']['hypothesis']
    return texts
```

```python
def evaluate_bertscore_avg(candidates1, references1, candidates2, references2):
    P_1, R_1, F1_1 = score(candidates1, references1, lang="en", rescale_with_baseline=True)
    P_2, R_2, F1_2 = score(candidates2, references2, lang="en", rescale_with_baseline=True)
    return 0.5*(P_1.mean().item() + P_2.mean().item()), 0.5*(R_1.mean().item() + R_2.mean().item()), 0.5*(F1_1.mean().item() + F1_2.mean().item())
```

```python
def evaluate_bleu_avg(candidates1, references1, candidates2, references2):
    scores1 = [sentence_bleu([ref], cand) for cand, ref in zip(candidates1, references1)]
    scores2 = [sentence_bleu([ref], cand) for cand, ref in zip(candidates2, references2)]
    return 0.5*(np.mean(scores1) + np.mean(scores2))
```

```python
def evaluate_meteor_avg(candidates1, references1, candidates2, references2):
    scores1 = []
    for reference, candidate in zip(references1, candidates1):
        # Tokenize the reference and hypothesis
        reference_tokens = nltk.word_tokenize(reference)
        candidate_tokens = nltk.word_tokenize(candidate)
        
        # Calculate METEOR score
        score = meteor_score([reference_tokens], candidate_tokens)
        scores1.append(score)
    
    scores2 = []
    for reference, candidate in zip(references2, candidates2):
        # Tokenize the reference and hypothesis
        reference_tokens = nltk.word_tokenize(reference)
        candidate_tokens = nltk.word_tokenize(candidate)
        
        # Calculate METEOR score
        score = meteor_score([reference_tokens], candidate_tokens)
        scores2.append(score)
    
    return 0.5*(np.mean(scores1) + np.mean(scores2))
```

```python
def evaluate_rouge_avg(candidates1, references1, candidates2, references2):
    rouge = Rouge()
    scores1 = []
    for cand, ref in zip(candidates1, references1):
        # Ensure each reference is a list of strings for compatibility with the ROUGE package
        score = rouge.get_scores(cand, ref)
        # Extract the F1 score from the ROUGE-L metric
        rouge_l_f1 = score[0]['rouge-l']['f']
        scores1.append(rouge_l_f1)
    
    scores2 = []
    for cand, ref in zip(candidates2, references2):
        # Ensure each reference is a list of strings for compatibility with the ROUGE package
        score = rouge.get_scores(cand, ref)
        # Extract the F1 score from the ROUGE-L metric
        rouge_l_f1 = score[0]['rouge-l']['f']
        scores2.append(rouge_l_f1)
    
    return 0.5*(np.mean(scores1) + np.mean(scores2))
```
